```python
import pandas as pd
import numpy as np
```

```python
data = {
    "Name": ["Ali", "Sara", "Ahmed", "Fatima", "Usman"],
    "Age": [22, 25, 24, np.nan, 27],
    "City": ["Lahore", "Karachi", "Islamabad", "Lahore", "Karachi"],
    "Salary": [50000, 60000, 55000, 58000, np.nan]
}

df = pd.DataFrame(data)

df.to_csv("data.csv", index=False)

print("CSV file created")
```

```python
df = pd.read_csv("data.csv")
print(df)
```

```python
df.to_excel("data.xlsx", index=False)

df_excel = pd.read_excel("data.xlsx")

print(df_excel)
```

```python
df.to_json("data.json")

df_json = pd.read_json("data.json")

print(df_json)
```

```python
print("Head:\n", df.head())
print("Tail:\n", df.tail())
print("Info:")
df.info()
print("Describe:\n", df.describe())
print("Shape:", df.shape)
```

```python
# Select single column
print(df["Name"])

# Select multiple columns
print(df[["Name", "Salary"]])

# Filter rows
print(df[df["Age"] > 23])
```

```python
# Sort by one column
print(df.sort_values(by="Salary"))

# Sort by multiple columns
print(df.sort_values(by=["City", "Salary"]))
```

```python
print(df.isnull())

# Drop missing values
print(df.dropna())

# Fill missing values
df["Age"].fillna(df["Age"].mean(), inplace=True)
df["Salary"].fillna(df["Salary"].mean(), inplace=True)

print(df)
```

```python
# Add new column
df["Bonus"] = df["Salary"] * 0.10

# Rename column
df.rename(columns={"Salary": "Monthly_Salary"}, inplace=True)

# Delete column
df.drop("City", axis=1, inplace=True)

print(df)
```
