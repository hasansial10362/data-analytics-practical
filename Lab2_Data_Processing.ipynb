```python
import pandas as pd

data = {
    "Name": ["Ali", "Sara", "Ahmed", "Fatima"],
    "Math": [85, 90, 78, 92],
    "Science": [88, 76, 85, 95]
}

df = pd.DataFrame(data)

df.to_csv("students.csv", index=False)

print("CSV file created")
```

```python
def calculate_average(marks):
    return sum(marks) / len(marks)

marks = [85, 90, 78, 92]

print("Average:", calculate_average(marks))
```
