```python
import numpy as np
```

```python
# 1D array
arr1 = np.array([1, 2, 3, 4])
print("1D Array:", arr1)
# 2D array
arr2 = np.array([[1, 2], [3, 4]])
print("2D Array:\n", arr2)

# 3D array
arr3 = np.array([[[1, 2], [3, 4]]])
print("3D Array:\n", arr3)
```

```python
arr = np.array([10, 20, 30, 40])

print(arr[0])      # first element
print(arr[1:3])    # slicing
```

```python
arr = np.array([1,2,3,4,5,6])

reshaped = arr.reshape(2,3)

print(reshaped)
```

```python
a = np.array([1,2,3])
b = np.array([4,5,6])

print("Addition:", a + b)
print("Multiplication:", a * b)
```

```python
arr = np.array([1,2,3])

print(arr + 5)
```

```python
arr = np.array([10,20,30,40])

print("Mean:", np.mean(arr))
print("Median:", np.median(arr))
print("Std Dev:", np.std(arr))
print("Percentile:", np.percentile(arr, 50))
```

```python
a = np.array([[1,2],[3,4]])
b = np.array([[5,6],[7,8]])

print("Dot Product:", np.dot(a,b))
print("Matrix Multiplication:", np.matmul(a,b))
```

```python
# random numbers
rand = np.random.rand(3)
print(rand)

# random integers
rand_int = np.random.randint(1,10,5)
print(rand_int)
```

```python
temps = np.array([30, 32, 35, 31, 29, 28, 33])

print("Average Temp:", np.mean(temps))
print("Max Temp:", np.max(temps))
print("Min Temp:", np.min(temps))
```

```python
import time

# Python list
start = time.time()
lst = [i for i in range(1000000)]
lst = [i+5 for i in lst]
print("List Time:", time.time() - start)

# NumPy
start = time.time()
arr = np.arange(1000000)
arr = arr + 5
print("NumPy Time:", time.time() - start)
```
