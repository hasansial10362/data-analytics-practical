```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = {
    "Day": [1,2,3,4,5,6,7],
    "Temperature": [30,32,35,31,29,28,33],
    "Cases": [100,150,200,180,120,90,160],
    "Humidity": [60,65,70,68,72,75,69]
}

df = pd.DataFrame(data)
print(df)
```

```python
plt.plot(df["Day"], df["Temperature"])
plt.title("Temperature Over Days")
plt.xlabel("Day")
plt.ylabel("Temperature")
plt.show()
```

```python
plt.bar(df["Day"], df["Cases"])
plt.title("COVID Cases")
plt.xlabel("Day")
plt.ylabel("Cases")
plt.show()
```

```python
plt.hist(df["Temperature"])
plt.title("Temperature Distribution")
plt.show()
```

```python
plt.scatter(df["Temperature"], df["Cases"])
plt.title("Temperature vs Cases")
plt.xlabel("Temperature")
plt.ylabel("Cases")
plt.show()
```

```python
sns.boxplot(data=df["Temperature"])
plt.title("Box Plot of Temperature")
plt.show()
```

```python
sns.violinplot(data=df["Temperature"])
plt.title("Violin Plot of Temperature")
plt.show()
```

```python
corr = df.corr()
sns.heatmap(corr, annot=True)
plt.title("Correlation Heatmap")
plt.show()
```

```python
plt.plot(df["Day"], df["Temperature"], color="red", label="Temp")
plt.title("Customized Plot")
plt.legend()
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(df["Day"], df["Temperature"])
plt.title("Temp")

plt.subplot(1,2,2)
plt.plot(df["Day"], df["Cases"])
plt.title("Cases")

plt.show()
```
